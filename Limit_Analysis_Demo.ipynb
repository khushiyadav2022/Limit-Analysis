```python
# @title
import pandas as pd
from datetime import datetime, timedelta
from tabulate import tabulate  # Import the tabulate function


available_domains = [
    "xyz.com","abc.io","yyy.com","bbb.com","aaa.co","hjk.us",
   "cvc.io","ddd.us","fff.us","yyy.net","hhh.com",
    "gfds.com","rrr.net","ggg.com","kgs.us","drm.net",
    "gdr.com","hry.net"
]

# Get user input for start date, end date, and domains
start_date_str = input("Enter the start date (MM/DD/YYYY): ")
end_date_str = input("Enter the end date (MM/DD/YYYY): ")

start_date = datetime.strptime(start_date_str, '%m/%d/%Y')
end_date = datetime.strptime(end_date_str, '%m/%d/%Y')

num_domains = int(input("Enter the number of domains: "))
domains = []

for i in range(num_domains):
    print("Select a domain name:")
    for index, domain_name in enumerate(available_domains, start=1):
        print(f"{index}. {domain_name}")

    domain_index = int(input(f"Enter the number for domain {i+1}: ")) - 1
    domain_name = available_domains[domain_index]

    current_limit = int(input(f"Enter current limit for {domain_name}: "))
    warm_up_activity = int(input(f"Enter warm up activity for {domain_name}: "))
    hike = float(input(f"Enter hike for {domain_name}: "))

    domain_info = {
        "name": domain_name,
        "current_limit": current_limit,
        "warm_up_activity": warm_up_activity,
        "hike": hike
    }

    domains.append(domain_info)

# Rest of your code remains unchanged
delta = timedelta(days=1)
data = []

current_limits = {domain["name"]: domain["current_limit"] for domain in domains}

while start_date <= end_date:
    formatted_date = start_date.strftime('%m/%d/%Y')

    for domain_info in domains:
        domain = domain_info["name"]
        current_limit = current_limits[domain]
        warm_up_activity = domain_info["warm_up_activity"]
        hike = domain_info["hike"]
        total = current_limit + warm_up_activity
        column_f_times_hike = total * hike

        next_limit = current_limit + column_f_times_hike

        data.append([
            formatted_date,
            domain,
            current_limit,
            warm_up_activity,
            hike,
            total,
            column_f_times_hike,
            next_limit
        ])

        current_limits[domain] = next_limit

    start_date += delta

columns = [
    "Date", "Domain", "Current DB Limit", "Warm Up Activity",
    "Hike", "Current DB Limit+ Warm Up Activity",
    "Column F * Hike", "Next Day Limit"
]


# df1 = pd.DataFrame(data, columns=columns).round()
df1 = pd.DataFrame(data, columns=columns)
# print("Initial DataFrame (df1):")
# print(df1.head(50))


# Print the initial DataFrame (df1) in a tabular format
print("Initial DataFrame (df1):")
print(tabulate(df1.head(100), headers='keys', tablefmt='psql'))  # Use the tabulate function

# # Display the interactive DataFrame using HTML formatting
# display(HTML(df1.to_html(index=False, escape=False)))




# ... Existing code ...

# Ask the user if they want to make changes to the DataFrame
change_df = input("Do you want to make changes to this DataFrame? (yes/no): ")

# ... Existing code ...

# ... Your existing code ...

if change_df.lower() == "yes":
    # Get user input for new start date
    new_start_date_str = input("Enter the new start date (MM/DD/YYYY): ")
    new_start_date = datetime.strptime(new_start_date_str, '%m/%d/%Y')

    # Convert the 'Date' column in df1 to datetime objects
    df1['Date'] = pd.to_datetime(df1['Date'], format='%m/%d/%Y')

    # Get user input for new current limit, warm up activity, and hike
    new_current_limit = int(input("Enter new current limit: "))
    new_warm_up_activity = int(input("Enter new warm up activity: "))
    new_hike = float(input("Enter new hike: "))

    # Create a new DataFrame df2 with updated values
    df2_data = []
    current_limits = {domain["name"]: domain["current_limit"] for domain in domains}

    delta = timedelta(days=1)

    while new_start_date <= end_date:
        for domain_info in domains:
            domain = domain_info["name"]
            new_current_limit = current_limits[domain]
            new_warm_up_activity = domain_info["warm_up_activity"]
            new_hike = domain_info["hike"]
            new_total = new_current_limit + new_warm_up_activity
            new_column_f_times_hike = new_total * new_hike
            new_next_limit = new_current_limit + new_column_f_times_hike

            df2_data.append([
                new_start_date,
                domain,
                new_current_limit,
                new_warm_up_activity,
                new_hike,
                new_total,
                new_column_f_times_hike,
                new_next_limit
            ])

            current_limits[domain] = new_next_limit

        new_start_date += delta



    # Convert the df2_data into a DataFrame
    df2 = pd.DataFrame(df2_data, columns=columns)

    # Filter data from df1 up to the day before the new start date
    filtered_df1 = df1[df1['Date'] < new_start_date_str]


    # Create a copy of df2 and adjust the start date
    df2_copy = df2.copy()
    df2_copy['Date'] = df2_copy['Date'] + pd.to_timedelta((new_start_date - start_date).days, unit='D')

    # # Filter data from df2 from new start date onwards
    # filtered_df2 = df2[df2['Date'] >= new_start_date]

    # Create a dataframe with a single row containing dashes
    separator_row = pd.DataFrame({'Date': ['-------------'], 'Domain': ['-------------'],
                              'Current DB Limit': ['-------------'], 'Warm Up Activity': ['-------------'],
                              'Hike': ['-------------'], 'Current DB Limit+ Warm Up Activity': ['-------------'],
                              'Column F * Hike': ['-------------'], 'Next Day Limit': ['-------------']})

    # Combine the filtered df1 and adjusted df2
    df_combined = pd.concat([filtered_df1,separator_row,df2_copy], ignore_index=True)

    print("\nDataFrame after changes (df_combined):")
    # print(df_combined.head(100))
    print(tabulate(df_combined.head(100), headers='keys', tablefmt='psql'))

else:
    print("No changes made to the DataFrame.")
    print(tabulate(df1.head(100), headers='keys', tablefmt='psql'))
```
